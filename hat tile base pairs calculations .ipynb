```python
#intial values 
r=0.34 #
D=2.7 #interhelical distance 
alpha=60 #angle end of base sub unit
l1= 8 #shortest string 
H=16 #number of helices per subunit

E= D/(m.tan(m.radians(alpha/2)))
n_bp=D/(r*m.tan(m.radians(alpha/2)))
print(E, n_bp)

#number of base pairs in each string 
string_lengths=[]
for j in range(1,17):
    if j<=8:
        length_string= round(l1+((j-1)*n_bp))
        string_lengths.append(length_string)
    if j>=9:
        length_string= round(l1+((H-j)*n_bp))
        string_lengths.append(length_string)
print(string_lengths, max(string_lengths))

string_ls=sum(string_lengths)
print(string_ls)
total_strings=string_ls*8
print(total_strings)
total=7249

total-total_strings

gap= D/(m.tan(m.radians(alpha/2)))

h=(max(string_lengths)-1)*r+gap 
h_nt=h/r
print(h,h_nt, n_bp)
```

```python
#scaffold strand coordinates
pc=15
dhtd=10.44
ddh=2
r=0.34 #
D=2.69  #interhelical distance 
alpha=60 #angle end of base sub unit
l1= 8 #shortest string 
H=16 #number of helices per subunit

scaffold_coords=[]
for n in range(0,len(string_lengths)):
    j=n+1
    coords=[]
    
    for k in range(0,string_lengths[n]):
        i=k+1
        
        x=-h+(0.34*(i-1))
        
        if j%2 ==1:
            a1=(D-2)/2
            b1=((H/2)-j)*D
            theta=((360*(i-1))/10.44)+15
            c1=(1-m.cos(m.radians(theta)))
            
            
            y=a1+b1+c1
            z=-(m.sin(m.radians(theta)))
        else:
            y=(-(D-ddh)/ddh)+((H/2)-j+1)*D -(1-m.cos(m.radians(((360*(i-1))/dhtd)-pc)))
            z=-m.sin(m.radians(((360*(i-1))/dhtd)-15))
        pos=(x,y,z)
        
        scaffold_coords.append(pos)
    
print(scaffold_coords)
    

```

```python
#graphing coords
x_coords=[]
y_coords=[]
for i in range(0,len(scaffold_coords)):
    point= scaffold_coords[i]
    x=point[0]
    y=point[1]
    x_coords.append(x)
    y_coords.append(y)

plt.plot(x_coords,y_coords,".", markersize="1")
plt.plot(x_coords[449],y_coords[449],"x")
plt.plot(0,0,"x")


```

```python
#applying a transformation to
len(x_coords)/2
scaffold_coords_trans=[]

for i in range(0, len(x_coords)):
    if i<=449:
        point=scaffold_coords[i]
        x=point[0]
        y=point[1]
        z=point[2]
    else:
        point=scaffold_coords[i]
        beta=m.degrees(m.atan(point[1]/point[0]))
        theta=(180+beta)-(alpha)
        rad=m.sqrt((point[0])**2+(point[1])**2)
        x=rad*m.cos(m.radians(theta))
        y=rad*m.sin(m.radians(theta))
        z=point[2]
   
    pos=(x,y,z)
    
    scaffold_coords_trans.append(pos)

print(scaffold_coords_trans)
print(beta)
```

```python
x_coords=[]
y_coords=[]
for i in range(0,len(scaffold_coords_trans)):
    point= scaffold_coords_trans[i]
    x=point[0]
    y=point[1]
    x_coords.append(x)
    y_coords.append(y)

plt.plot(x_coords,y_coords,".", markersize="1")
plt.axis('equal')
kite_x=(0,-h,-h,-h*m.cos(m.radians(alpha)),0)
kite_y=(0,0,h*m.tan(m.radians(alpha/2)),h*m.sin(m.radians(alpha)),0)
        
plt.plot(kite_x,kite_y, color="grey")

plt.plot(x_coords[29],y_coords[29],"x")

```
